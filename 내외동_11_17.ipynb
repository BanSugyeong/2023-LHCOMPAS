# 내외동 데이터

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/공모전/우리동네 문제해결/기온강수/내외동_1시간기온강수_20211001_20220930.csv')
df
```

11:00

```python
df_11 = df[df['hour'].isin([1100])]
df_11
```

```python
df_1100 = df_11.groupby(['year', 'month', 'day', 'hour']).agg(c_avr=('c', 'mean'), p_avr=('p', 'mean')).reset_index()
df_1100.to_csv('내외동_11.csv', encoding = 'cp949', index = False)
df_1100
```

17:00

```python
df_17 = df[df['hour'].isin([1700])]
df_17
```

```python
df_1700 = df_17.groupby(['year', 'month', 'day', 'hour']).agg(c_avr=('c', 'mean'), p_avr=('p', 'mean')).reset_index()
df_1700.to_csv('내외동_17.csv', encoding = 'cp949', index = False)
df_1700
```
