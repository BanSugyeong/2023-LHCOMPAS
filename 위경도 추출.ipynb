```python
import pandas as pd
df = pd.read_csv('장애인전국체전_주소.csv',encoding='cp949')
df
```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent = 'South Korea')

def geocoding_lat(address):
    try:
        geo = geo_local.geocode(address)
        return geo.latitude
    
    except:
        return 0
    
    
def geocoding_lon(address):
    try:
        geo = geo_local.geocode(address)
        return geo.longitude
    
    except:
        return 0
    
df['위도'] = df['주소'].apply(geocoding_lat)
df['경도'] = df['주소'].apply(geocoding_lon)
df
```

```python
df.to_csv('장애인전국체전_위경도.csv', encoding = 'cp949', index = False)
```
