## 유동인구 데이터 2022.04 ~ 2022.09

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('유동인구데이터_22.04.csv', encoding='cp949', index_col=0)
df
```

## 월별로 분리

```python
month_list = [4, 5, 6, 7, 8, 9]
result_list = []

for month in month_list:
    result = df[df['기준_월'].astype(str).str.contains(str(month), na=False)]
    result_list.append(result)
```

```python
month_4 = result_list[0]
month_4.to_csv('유동인구데이터_22년4월.csv', encoding='cp949', index=False)
month_4
```
