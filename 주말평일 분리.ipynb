```python
import datetime
import pandas as pd
import numpy as np

def split_weekend_weekday(start_date, end_date):
    weekend_data = []
    weekday_data = []

    date = start_date
    while date <= end_date:
        if date.weekday() < 5:  # 0은 월요일, 4는 금요일을 의미합니다.
            weekday_data.append(date.strftime("%d"))
        else:
            weekend_data.append(date.strftime("%d"))
        date += datetime.timedelta(days=1)

        # 다음달 1일로 이동
        if date.day == 1:
            date = date.replace(day=1, month=date.month + 1)

    return weekend_data, weekday_data

# 시작 날짜와 종료 날짜 설정
start_date = datetime.datetime(2021, 10, 1)
end_date = datetime.datetime(2021, 10, 31)

weekend, weekday = split_weekend_weekday(start_date, end_date)

print("주말 데이터:", weekend)
print("평일 데이터:", weekday)


#주말 데이터: ['2021-10-02', '2021-10-03', '2021-10-09', '2021-10-10', '2021-10-16', '2021-10-17', '2021-10-23', '2021-10-24', '2021-10-30', '2021-10-31']
#평일 데이터: ['2021-10-01', '2021-10-04', '2021-10-05', '2021-10-06', '2021-10-07', '2021-10-08', '2021-10-11', '2021-10-12', '2021-10-13', '2021-10-14', '2021-10-15', '2021-10-18', '2021-10-19', '2021-10-20', '2021-10-21', '2021-10-22', '2021-10-25', '2021-10-26', '2021-10-27', '2021-10-28', '2021-10-29']
```
