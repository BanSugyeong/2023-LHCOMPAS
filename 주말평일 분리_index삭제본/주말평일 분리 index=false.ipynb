```python
import datetime
import pandas as pd
import numpy as np
import pandas as pd
import os
import csv
```

```python
def split_weekend_weekday(start_date, end_date):
    weekend_data = []
    weekday_data = []

    date = start_date
    while date <= end_date:
        if date.weekday() < 5:  # 0은 월요일, 4는 금요일을 의미합니다.
            weekday_data.append(date.strftime("%d"))
        else:
            weekend_data.append(date.strftime("%d"))
        date += datetime.timedelta(days=1)

        # 다음달 1일로 이동
        if date.day == 1:
            date = date.replace(day=1, month=date.month + 1)

    return weekend_data, weekday_data
```

```python
# 시작 날짜와 종료 날짜 설정
start_date = datetime.datetime(2021, 10, 1)
end_date = datetime.datetime(2021, 10, 31)

weekend, weekday = split_weekend_weekday(start_date, end_date)

# 일 수를 1자리 숫자로 변경
weekend = [WE[:0] + str(int(WE[0:])) for WE in weekend]
weekday = [WD[:0] + str(int(WD[0:])) for WD in weekday]


print("주말 데이터:", weekend)
print("평일 데이터:", weekday)

weekend = [2, 3, 9, 10, 16, 17, 23, 24, 30, 31]
weekday = [1, 4, 5, 6, 7, 8, 11, 12, 13, 14, 15, 18, 19, 20, 21, 22, 25, 26, 27, 28, 29]

print("주말 데이터:", weekend)
print("평일 데이터:", weekday)
```
