## 유동인구 데이터 2021.10 ~ 2022.03

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('floating_population.csv', encoding='cp949', index_col=0)
df
```

## 월별로 분리

```python
month_list = [10, 11, 12, 1, 2, 3]
result_list = []

for month in month_list:
    result = df[df['기준_월'].astype(str).str.contains(str(month), na=False)]
    result_list.append(result)
```

```python
month_10 = result_list[0]
month_10.to_csv('유동인구데이터_21년10월.csv', encoding='cp949', index=False)
month_10
```

```python
month_11 = result_list[1]
month_11.to_csv('유동인구데이터_21년11월.csv', encoding='cp949', index=False)
month_11
```
