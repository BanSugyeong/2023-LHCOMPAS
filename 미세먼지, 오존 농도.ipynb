## 미세먼지, 오존 농도 현황

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('경상남도 김해시_(초)미세먼지 및 오존 농도 현황_20220731.csv', encoding='cp949')
df
```

## 오존 농도 분류

```python
def classify_ozone(df):
    # 0 = 좋음, 1 = 보통, 2 = 나쁨, 3 = 매우 나쁨
    for i, row in df.iterrows():
        levels = row['오존(O3)']
        if levels <= 0.030:
            df.at[i, '오존(O3)'] = 0
        elif levels > 0.030 or levels <= 0.09:
            df.at[i, '오존(O3)'] = 1
        elif levels > 0.09 or levels <= 0.150:
            df.at[i, '오존(O3)'] = 2
        else:
            df.at[i, '오존(O3)'] = 3
    return df

ozone = classify_ozone(df)

for index, row in ozone.iterrows():
    print(row)
```

## 미세먼지 농도 분류

```python
def classify_finedust(df):
    # 0 = 좋음, 1 = 보통, 2 = 나쁨, 3 = 매우 나쁨
    for i, row in df.iterrows():
        levels = row['미세먼지(PM10)']
        if levels <= 30:
            df.at[i, '미세먼지(PM10)'] = 0
        elif levels > 30 or levels <= 80:
            df.at[i, '미세먼지(PM10)'] = 1
        elif levels > 80 or levels <= 150:
            df.at[i, '미세먼지(PM10)'] = 2
        else:
            df.at[i, '미세먼지(PM10)'] = 3
    return df

finedust = classify_finedust(df)

for index, row in finedust.iterrows():
    print(row)
```

## 초 미세먼지 농도 분류

```python
def classify_ultra_finedust(df):
    # 0 = 좋음, 1 = 보통, 2 = 나쁨, 3 = 매우 나쁨
    for i, row in df.iterrows():
        levels = row['초미세먼지(PM2점5)']
        if levels <= 15:
            df.at[i, '초미세먼지(PM2점5)'] = 0
        elif levels > 15 or levels <= 35:
            df.at[i, '초미세먼지(PM2점5)'] = 1
        elif levels > 35 or levels <= 75:
            df.at[i, '초미세먼지(PM2점5)'] = 2
        else:
            df.at[i, '초미세먼지(PM2점5)'] = 3
    return df

ultra_finedust = classify_ultra_finedust(df)

for index, row in ultra_finedust.iterrows():
    print(row)
```
